```python
socioeconomic = socioeconomic.drop(socioeconomic[socioeconomic["year"]<=1970].index)
socioeconomic['country_year'] = socioeconomic[['country','year']].apply(lambda x : '{} {}'.format(x[0],x[1]), axis=1)
duplicatesocioeconomic = socioeconomic[socioeconomic.duplicated()]
display(duplicatesocioeconomic)
socioeconomic.sort_values(by=['country_year'], inplace=True)
socioeconomic_final=socioeconomic.reset_index()
display(socioeconomic_final)
socioeconomic_final['country_year'].value_counts()
```

```python
import requests
from bs4 import BeautifulSoup
res = requests.get("http://hdr.undp.org/en/content/health-index")
soup = BeautifulSoup(res.content,'html.parser')

display(soup)
```

```python

rows=soup.select('tr')
data = soup.select('td')
print (len(rows))
print (len(data))
print(data[1].text)
print(data[2].text)
type(rows[1])
print(rows)
```

```python
data
```

```python
resultado = []
for i in range(199):
    q=rows[i+1].select('td')
    
    filas={
            "id": str(q[0].text),
            "country":str(q[1].text),
            "1980": str(q[2].text).strip(),
            "1985": str(q[3].text).strip(),
            "1990": str(q[4].text).strip(),
            "2000": str(q[5].text).strip(),
            "2005": str(q[6].text).strip(),
            "2006": str(q[7].text).strip(),
            "2007": str(q[8].text).strip(),
            "2008": str(q[9].text).strip(),
            "2009": str(q[10].text).strip(),
            "2010": str(q[11].text).strip(),
            "2011": str(q[12].text).strip(),
            "2012": str(q[13].text).strip(),
            "2013": str(q[14].text).strip()
            }
    resultado.append(filas)
resultado
```

```python
health_index = pd.DataFrame(resultado)
health_index.sort_values(by=['country'], inplace=True)
display(health_index)
```

```python
health_index.drop(['id','1985','2005','2006','2007','2008','2009','2011','2012','2013'], axis=1, inplace=True)
health_index.head()
```

```python
health_index_melted= pd.melt(health_index, id_vars=['country'], value_vars=['1980', '1990','2000', '2010'], var_name='Year', value_name='health')
health_index_melted.sort_values(by=['country'], inplace=True)
health_index_melted['Year'] = health_index_melted['Year'].apply(pd.to_numeric) 
health_index_melted['country_year'] = health_index_melted[['country','Year']].apply(lambda x : '{} {}'.format(x[0],x[1]), axis=1)
duplicatehealth_index_melted = health_index_melted[health_index_melted.duplicated()]
display(duplicatehealth_index_melted)
health_index_melted.sort_values(by=['country_year'], inplace=True)
health_index_melted.drop(['country','Year'], axis=1, inplace=True)
health_final=health_index_melted.reset_index()
display(health_final)
```

```python
lista_health=[]
for i in health_final.country_year:
    lista_health.append(i.lstrip())
#print(lista_health)
health_final.country_year=lista_health
lista_socio=[]
for e in socioeconomic_final.country_year:
    lista_socio.append(e) 
#print(lista_socio)

for i in lista_health:
    if i not in lista_socio:
        lista=(i)
        print(lista)
```
